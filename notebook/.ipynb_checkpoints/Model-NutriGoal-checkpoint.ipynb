## Model Building

### BMR and Daily Calorie Requirement Calculation

```python
# Fungsi untuk menghitung BMR
def calculate_bmr(weight, height, age, gender):
    if gender.lower() == 'pria':
        return 66.5 + (13.75 * weight) + (5.003 * height) - (6.75 * age)
    else:
        return 655.1 + (9.563 * weight) + (1.850 * height) - (4.676 * age)

# Fungsi untuk menghitung kebutuhan kalori harian berdasarkan aktivitas
def calculate_daily_calories(bmr, activity_level):
    factors = {
        'sedentary': 1.2,
        'lightly active': 1.375,
        'moderately active': 1.55,
        'very active': 1.725
    }
    return bmr * factors[activity_level]

# Contoh data pengguna
user_data = {
    'weight': 70,  # kg
    'height': 175,  # cm
    'age': 25,
    'gender': 'pria',
    'activity_level': 'moderately active'
}

# Hitung BMR dan kebutuhan kalori harian
bmr = calculate_bmr(user_data['weight'], user_data['height'], user_data['age'], user_data['gender'])
daily_calories = calculate_daily_calories(bmr, user_data['activity_level'])

print(f"BMR: {bmr:.2f} kcal")
print(f"Kebutuhan kalori harian: {daily_calories:.2f} kcal")
```

### Nutritional Analysis and Food Recommendations

```python
# Pilih kolom nutrisi yang relevan
nutritional_columns = ['calories', 'protein (g)', 'carbohydrate (g)', 'fiber (g)', 'sugars (g)', 
                       'total_fat (g)', 'saturated_fat (g)', 'cholesterol (mg)', 'sodium (mg)', 
                       'vitamin_c (mg)', 'calcium (mg)', 'iron (mg)', 'potassium (mg)']
food_data = df[nutritional_columns]

# Preprocessing
food_data = food_data.fillna(0)
scaler = MinMaxScaler()
food_data_scaled = scaler.fit_transform(food_data)

# Split data menjadi training dan testing
train_data, val_data = train_test_split(food_data_scaled, test_size=0.3, random_state=42)

# Membuat model rekomendasi berbasis jaringan saraf
model = Sequential()
model.add(Dense(32, activation='relu', input_dim=train_data.shape[1]))
model.add(Dense(16, activation='relu'))
model.add(Dense(train_data.shape[1], activation='linear'))
model.compile(optimizer='adam', loss='mean_squared_error')

# Latih model
history1 = model.fit(train_data, train_data, epochs=20, batch_size=32, validation_data=(val_data, val_data))

# Visualisasi hasil pelatihan
def plot_loss(history):
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

plot_loss(history1)

# Definisikan daily_calories atau input pengguna lainnya
daily_calories = bmr  # Contoh input dari pengguna

# Sesuaikan user_input dengan fitur yang di-fit oleh MinMaxScaler
user_input_dict = {
    'calories': daily_calories,
    'protein (g)': 50,
    'carbohydrate (g)': 20,
    'fiber (g)': 10,
    'sugars (g)': 30,
    'total_fat (g)': 15,
    'saturated_fat (g)': 5,
    'cholesterol (mg)': 0,
    'sodium (mg)': 1500,
    'vitamin_c (mg)': 60,
    'calcium (mg)': 100,
    'iron (mg)': 18,
    'potassium (mg)': 3500
}

# Konversi input pengguna menjadi DataFrame
user_input_df = pd.DataFrame([user_input_dict])

# Sesuaikan urutan kolom dengan `scaler.feature_names_in_`
user_input_df = user_input_df[scaler.feature_names_in_]

# Lakukan scaling pada input pengguna
input_scaled = scaler.transform(user_input_df)

similarity_scores = cosine_similarity(input_scaled, food_data_scaled)

# Ambil rekomendasi makanan teratas
top_indices = np.argsort(similarity_scores, axis=1)[0][::-1][:5]
recommended_foods = df.iloc[top_indices]
print(recommended_foods[['name', 'calories', 'protein (g)', 'carbohydrate (g)', 'fiber (g)', 'sugars (g)', 
                       'total_fat (g)', 'saturated_fat (g)', 'cholesterol (mg)', 'sodium (mg)', 
                       'vitamin_c (mg)', 'calcium (mg)', 'iron (mg)', 'potassium (mg)']])
```

### Food Recommendations Based on User Preferences

**Preference Data Collection Users**

```python
# # Input preferensi pengguna
# user_preferences = ['chicken', 'rice', 'vegetables']

# # Filter data berdasarkan preferensi makanan
# def filter_by_preference(df, preferences):
#     return df[df['name'].str.contains('|'.join(preferences), case=False)]

# filtered_data = filter_by_preference(df, user_preferences)

# Input preferensi pengguna
user_preferences = ['chicken', 'rice', 'vegetables']

# Preprocessing data dengan TfidfVectorizer
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['name'])

# Membuat representasi preferensi pengguna
user_query = ' '.join(user_preferences)
user_vector = vectorizer.transform([user_query])

# Mendapatkan skor kesamaan antara preferensi pengguna dan dataset
similarity_scores = cosine_similarity(user_vector, tfidf_matrix)
df['similarity'] = similarity_scores.flatten()

# Menyaring data untuk meningkatkan kualitas rekomendasi berdasarkan preferensi
filtered_data = df[df['similarity'] > 0.1]
print(filtered_data.head())
filtered_data.head()
```

**Training Recommendation Models Based on Preferences Users**

```python
# # NLP untuk analisis preferensi
# vectorizer = TfidfVectorizer()
# tfidf_matrix = vectorizer.fit_transform(df['name'])

# # Dapatkan kesamaan antara preferensi pengguna dan dataset
# user_query = ' '.join(user_preferences)
# user_vector = vectorizer.transform([user_query])
# similarity_scores = cosine_similarity(user_vector, tfidf_matrix)

# # Rekomendasikan makanan berdasarkan kesamaan
# top_indices = np.argsort(similarity_scores.flatten())[::-1][:5]
# recommendations = df.iloc[top_indices]
# print(recommendations[['name', 'calories', 'protein (g)', 'carbohydrate (g)', 'fiber (g)', 'sugars (g)', 
#                        'total_fat (g)', 'saturated_fat (g)', 'cholesterol (mg)', 'sodium (mg)', 
#                        'vitamin_c (mg)', 'calcium (mg)', 'iron (mg)', 'potassium (mg)']])
# recommendations

# Memilih fitur yang relevan untuk model TensorFlow
features = ['calories', 'protein (g)', 'carbohydrate (g)', 'fiber (g)', 'sugars (g)', 
                       'total_fat (g)', 'saturated_fat (g)', 'cholesterol (mg)', 'sodium (mg)', 
                       'vitamin_c (mg)', 'calcium (mg)', 'iron (mg)', 'potassium (mg)']
X = filtered_data[features]
y = filtered_data['similarity']

# Normalisasi fitur
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Membagi data menjadi training dan testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Membuat model TensorFlow
model = Sequential([
    Dense(64, input_shape=(X_train.shape[1],), activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error', metrics=['mae'])

# Training model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=1)

# Evaluasi model
loss, mae = model.evaluate(X_test, y_test)
print(f"Model Loss: {loss}, Mean Absolute Error: {mae}")

# Membuat rekomendasi berdasarkan model yang sudah dilatih
predictions = model.predict(X_test)
top_indices = np.argsort(predictions.flatten())[::-1][:5]
recommended_foods = filtered_data.iloc[top_indices]

# Menampilkan rekomendasi makanan
print("Top 5 Food Recommendations:")
print(recommended_foods[['name', 'calories', 'protein (g)', 'carbohydrate (g)', 'fiber (g)', 'sugars (g)', 
                       'total_fat (g)', 'saturated_fat (g)', 'cholesterol (mg)', 'sodium (mg)', 
                       'vitamin_c (mg)', 'calcium (mg)', 'iron (mg)', 'potassium (mg)']])
recommended_foods
```
